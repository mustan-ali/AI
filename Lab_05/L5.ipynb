### **Correlation**
- realation between two variables

```python
import matplotlib.pyplot as plt
import seaborn as sb

dataframe = sb.load_dataset("tips")
sb.pairplot(dataframe, kind="scatter")
plt.show()
```

```python
# Correlation using pandas

import pandas as pd

x = pd.Series(range(1, 20))
y = pd.Series([5, 45, 88, 3, 5, 44, 9, 62, 11, 50])

print(x.corr(y))
```

```python
# Correlation using numpy

import numpy as np

x = np.arange(0, 10)
y = np.array([5, 45, 88, 3, 5, 44, 9, 62, 11, 50])

find_correlation = np.corrcoef(x, y)

print("Value of X", x)
print("Value of Y", y)

print("\nCorrelation Coefficient\n", find_correlation)
```

### **K-Nearst Neighbors (KNN)**
- predict the value of a target variable by learning from the k-nearest data points in the feature space

```python
import pandas as pd
from sklearn.datasets import load_iris

iris_dataset = load_iris()

print(iris_dataset.feature_names)
print(iris_dataset.target_names)
```

```python
df = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
df.head()
```

```python
# Adding target column to the dataframe

df["target"] = iris_dataset.target
df.head()
```

```python
df["flower_name"] = df.target.apply(lambda x: iris_dataset.target_names[x])
df.head(10)
```

```python
df[df.target == 1].head()
```

```python
df0 = df[:50]   # Values from 0 to 49
df1 = df[50:100]    # Values from 50 to 99
df2 = df[100:]  # Values from 100 to 149

import matplotlib.pyplot as plt

# Sepal Length vs Sepal Width (Setosa vs Versicolor)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color="green", marker="+")
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color="blue", marker=".")

# Petal Length vs Petal Width (Setosa vs Versicolor)
# plt.xlabel("Petal Length")
# plt.ylabel("Petal Width")

# plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color="green", marker="+")
# plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color="blue", marker=".")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

x = df.drop(["target", "flower_name"], axis="columns")
y = df.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(x_train, y_train)

knn.score(x_test, y_test)
```

```python
predicted_value = knn.predict([[6.5, 2.8, 4.6, 1.5]])

print("KNN Species Prediction: ", predicted_value)
```
