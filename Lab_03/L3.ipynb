### **FIZZ BUZZ**
- Divisible by 3 FIZZ
- Divisible by 5 BUZZ
- Divisible by 3 & 5 FIZZ BUZZ

```python
number = int(input())

if number % 3 == 0:
    print("FIZZ")
if number % 5 == 0:
    print("BUZZ")
```

### **Filing**

- Text Mode (Default)
    - **x**: Create (Returns an error if the file exists) 
    - **r**: Read
    - **w**: Write  (Creates a new file or overwrites the existing file) 
    - **a**: Append (Creates a new file or appends to the existing file)
- Binary Mode


```python
f = open("abc.txt", "a")
f.write("Hello World")
f.close()

print("Data has been written to file.")
```

```python
filename = input("Enter Name of your file:")
open(filename + ".txt", "x")

print("File has been created")
```

```python
import array as arr

num = arr.array("i", np.arange(1, 11))

print(num)

f = open("test.txt", "a")

for i in num:
    i *= 2
    f.write(str(i) + "\n")
    print(i)

f.close()
```
